# 02 — Call Azure OpenAI from Python

**Goal**: Make the same chat completion call you make in C#, but in Python.

**Prerequisite**: Azure OpenAI resource from Week 2 + filled-in `.env` file.

```python
from dotenv import load_dotenv
import os

load_dotenv()

endpoint = os.getenv("AZURE_OPENAI_ENDPOINT")
api_key = os.getenv("AZURE_OPENAI_KEY")
deployment = os.getenv("AZURE_OPENAI_CHAT_DEPLOYMENT")

print("Endpoint configured:", endpoint is not None)
```

## Chat completion

```python
from openai import AzureOpenAI

client = AzureOpenAI(
    azure_endpoint=endpoint,
    api_key=api_key,
    api_version="2024-02-15-preview",
)

response = client.chat.completions.create(
    model=deployment,
    messages=[
        {"role": "system", "content": "You are a concise assistant."},
        {"role": "user", "content": "Explain RAG in one sentence."},
    ],
    temperature=0.3,
    max_tokens=120,
)

print(response.choices[0].message.content)
```

## Embeddings

```python
embed_deployment = os.getenv("AZURE_OPENAI_EMBED_DEPLOYMENT")

result = client.embeddings.create(
    model=embed_deployment,
    input=["The cat sat on the mat", "A feline rested on the rug"],
)

vec1 = result.data[0].embedding
vec2 = result.data[1].embedding
print(f"Vector dim: {len(vec1)}")
print(f"First 5 values: {vec1[:5]}")
```

## Cosine similarity (so you see embeddings work)

```python
import numpy as np

def cosine(a, b):
    a, b = np.array(a), np.array(b)
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

print("Similar sentences:", round(cosine(vec1, vec2), 4))

result2 = client.embeddings.create(
    model=embed_deployment,
    input=["The price of bitcoin"],
)
vec3 = result2.data[0].embedding
print("Unrelated:        ", round(cosine(vec1, vec3), 4))
```
