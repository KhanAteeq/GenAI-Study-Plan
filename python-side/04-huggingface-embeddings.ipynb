# 04 — Hugging Face: Local Embeddings

**Goal**: Generate embeddings locally (no API calls, no cost) and compare with Azure OpenAI embeddings.

**Why care**: Lots of GenAI tutorials use Hugging Face. Knowing how to swap models is a real skill.

First run downloads the model (~80 MB) — be patient.

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Small, fast, good-enough model
model = SentenceTransformer("all-MiniLM-L6-v2")
print("Model loaded. Vector dim:", model.get_sentence_embedding_dimension())
```

## Embed sentences

```python
sentences = [
    "The cat sat on the mat",
    "A feline rested on the rug",
    "Bitcoin price hit a new high",
]

vectors = model.encode(sentences)
print("Shape:", vectors.shape)   # (3, 384)
```

## Compare similarities

```python
def cosine(a, b):
    return float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))

print("cat/feline (similar):  ", round(cosine(vectors[0], vectors[1]), 4))
print("cat/bitcoin (different):", round(cosine(vectors[0], vectors[2]), 4))
```

## Trade-offs to remember

| | Azure OpenAI (text-embedding-3-small) | Hugging Face (MiniLM) |
|---|---|---|
| Vector dim | 1536 | 384 |
| Cost | $0.02 / 1M tokens | Free (runs locally) |
| Quality | Higher | Good for many uses |
| Latency | Network round-trip | Local, fast on CPU |
| Privacy | Data leaves your machine | Stays local |

For your **production app**: Azure OpenAI.
For **prototyping / sensitive data**: Hugging Face local.
