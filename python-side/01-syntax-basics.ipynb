## Variables (no types needed)

```python
name = "Atik"
age = 30
is_student = True
print(name, age, is_student)
```

## Lists, dicts, tuples

```python
fruits = ["apple", "banana", "cherry"]
print(fruits[0], len(fruits))

person = {"name": "Atik", "role": "engineer"}
print(person["name"])

point = (10, 20)  # immutable
print(point)
```

## Functions

```python
def greet(name: str, greeting: str = "Hello") -> str:
    return f"{greeting}, {name}!"

print(greet("Atik"))
print(greet("Atik", greeting="Hi"))
```

## Control flow (indentation = scope!)

```python
for i in range(5):
    if i % 2 == 0:
        print(f"{i} is even")
    else:
        print(f"{i} is odd")
```

## List comprehensions (Python's killer feature)

```python
squares = [x * x for x in range(10)]
print(squares)

evens = [x for x in range(20) if x % 2 == 0]
print(evens)
```

## Classes

```python
class Person:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def greet(self) -> str:
        return f"Hi, I'm {self.name}"

p = Person("Atik", 30)
print(p.greet())
```
